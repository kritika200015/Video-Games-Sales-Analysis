```python
df = pd.read_csv("D:/vgsales.csv")
df.head()
```

# cleaning the data


1. droping rows with NaN

```python
nan_df = df[df.isna().any(axis=1)]
nan_df.head()
```

```python
df = df.dropna(how='all')
df.head()
```

2. making infinite to nan and then removing them

```python
df.replace([np.inf, -np.inf], np.nan, inplace=True) 
df.dropna(inplace=True) 
df
```

3. making year as integer

```python
df['Year']= df['Year'].astype('int32')
df
```

4. making all sales types to float

```python
df['NA_Sales'] = pd.to_numeric(df['NA_Sales'])
df['EU_Sales'] = pd.to_numeric(df['EU_Sales'])
df['JP_Sales'] = pd.to_numeric(df['JP_Sales'])
df['Other_Sales'] = pd.to_numeric(df['Other_Sales'])
df['Global_Sales'] = pd.to_numeric(df['Global_Sales'])

df
```

5.Determining the best year of sales

```python
grouping = df.groupby('Year').sum()
```

```python
import matplotlib.pyplot as plt

Year = range(1,40)

plt.bar(Year,grouping['Global_Sales'])
plt.xticks(Year,rotation='vertical',size=7)
plt.xlabel("Years from 1980-2020")
plt.ylabel("Global_Sales")
plt.show()
```
